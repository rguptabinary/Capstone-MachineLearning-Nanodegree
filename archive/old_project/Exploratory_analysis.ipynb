```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from utilities import *
%matplotlib inline
```

```python
start_date = '2010-01-16'
end_date = '2014-01-16'
dates = pd.date_range(start_date, end_date)
```

### Below are the plots depicting the distribution and volatility of different stocks.

## SPY

```python
df_spy = get_data('SPY', dates)

ax = df_spy.plot(title='SPY',y=['Adj Close', 'Rolling mean'])
ax.set(xlabel='Date', ylabel='Price')
plt.show()
```
